```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
data={
    'Name':["Mr.Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William","Mr. James", "Mr. Timothy", "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite","Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    'Height': [1.80, 1.61, 1.78, 1.74, 1.55, 1.90, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    'Weight': [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    'AGE': [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]

}
df = pd.DataFrame(data)
```

## 12. Loai bo cac gia tri trung lap dua tren name, giu lai hang dau

```python
df.drop_duplicates(subset=["Name"], keep="first", inplace=True)
print(df)
```

## 13. Thay the AGE >40 bang gia tri trung binh cua AGE

```python
df.loc[df["AGE"] > 40, "AGE"] = df["AGE"].mean()
print(df)
```

## 14. Tinh chi so BMI

```python
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)
print(df)
```

## 15. Phan loai BMI

```python
def classify_bmi(bmi):
    if bmi < 18.5:
        return "0"  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return "1"  # Bình thường
    elif 23 <= bmi <= 24.9:
        return "2"  # Thừa cân
    else:
        return "3"  # Béo phì

df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)
print(df)
```

## 16. Chuan hoa cs_BMI

```python
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
print(df)
```

## 17. Xac dinh gioi tinh

```python
def extract_gender(name):
    if name.startswith("Mr."):
        return "male"
    else:
        return "female"

df["GT"] = df["Name"].apply(extract_gender)
df["Name"] = df["Name"].str.replace("Mr. |Mrs. |Miss. ", "", regex=True)
print(df)
```

## 18. Ve bieu do tron cho GT

```python
plt.figure()
df["GT"].value_counts().plot(kind="pie", autopct="%.1f%%")
plt.title("Biểu đồ tròn của GT")
plt.show()
```

## 19. Truc quan hoa du lieu Phan_loai

```python
plt.figure()
df[df["GT"] == "female"]["Phan_Loai"].value_counts().plot(kind="bar")
plt.title("Phân loại BMI của nữ")
plt.show()
```
