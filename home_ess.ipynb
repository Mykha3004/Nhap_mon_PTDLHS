```python
# Thống kê số lượng dữ liệu thiếu trong cột "DT"
missing_count = df["DT"].isna().sum()
print(f"Số lượng giá trị thiếu trong cột 'DT': {missing_count}")

# Điền giá trị thiếu bằng "Khác"
df["DT"].fillna("Khác", inplace=True)

# Hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```

```python


# Thống kê số lượng dữ liệu thiếu trong cột "T1"
missing_count_T1 = df["T1"].isna().sum()
print(f"Số lượng giá trị thiếu trong cột 'T1': {missing_count_T1}")

# Thay thế dữ liệu thiếu bằng giá trị trung bình của cột "T1"
mean_T1 = df["T1"].mean()
df["T1"].fillna(mean_T1, inplace=True)

# Hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```

```python
# Xác định các cột điểm số (giả sử các cột điểm số có tên bắt đầu bằng 'T', 'L', 'H', 'S', 'V', 'X', 'D', 'N', 'DH')
diem_columns = [col for col in df.columns if col.startswith(('T', 'L', 'H', 'S', 'V', 'X', 'D', 'N', 'DH'))]

# Thống kê và thay thế dữ liệu thiếu bằng giá trị trung bình
for col in diem_columns:
    missing_count = df[col].isna().sum()
    if missing_count > 0:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)
        print(f"Đã thay thế {missing_count} giá trị thiếu trong cột '{col}' bằng giá trị trung bình: {mean_value}")
        # Hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```

```python
# Tính trung bình môn cho từng năm học
for i in range(1, 4):  # Lớp 10, 11, 12 tương ứng với T1, T2, T3,...
    df[f"TBM{i}"] = (df[f"T{i}"] * 2 + df[f"L{i}"] + df[f"H{i}"] + df[f"S{i}"] + df[f"V{i}"] * 2 + df[f"X{i}"] + df[f"D{i}"] + df[f"N{i}"]) / 10
    #hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```

```python
# biến xếp loại XL1, XL2, XL3 dựa trên TBM1, TBM2, TBM3
for i in range(1, 4):
    df[f"XL{i}"] = df[f"TBM{i}"].apply(lambda x: "Y" if x < 5.0 else
                                        "TB" if x < 6.5 else
                                        "K" if x < 8.0 else
                                        "G" if x < 9.0 else "XS")

# Hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```

```python
# Chuyển đổi điểm từ thang 10 sang thang 4 bằng Min-Max Normalization
for i in range(1, 4):
    df[f"US_TBM{i}"] = (df[f"TBM{i}"] - df[f"TBM{i}"].min()) / (df[f"TBM{i}"].max() - df[f"TBM{i}"].min()) * 4

# Hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```

```python
# Tạo biến kết quả xét tuyển KQXT
def determine_kqxt(row):
    khoi_a_a1 = (row["DH1"] * 2 + row["DH2"] + row["DH3"]) / 4
    khoi_b = (row["DH1"] + row["DH2"] * 2 + row["DH3"]) / 4
    return 1 if khoi_a_a1 >= 5.0 or khoi_b >= 5.0 else 0

df["KQXT"] = df.apply(determine_kqxt, axis=1)

# Hiển thị DataFrame sau khi hiệu chỉnh
print(df.head())
```
