##  Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột “Name” chứa bao gồm “Firstname” và “Lastname”, giải pháp là ta sẽ tách ra làm 2 cột 

```python
import pandas as pd

# Đường dẫn đến file CSV
file_path = r"C:\Users\Admin\Documents\207CT28171\heart_disease_dataset.csv"

# Đọc dữ liệu từ tệp CSV, giả định rằng tệp không có header
# Chỉ định tên cột mới nếu cần
column_names = ["Name", "Age", "Sex", "ChestPainType", "RestingBP", "Cholesterol", "FastingBS", "RestingECG", "MaxHR", "ExerciseAngina", "Oldpeak", "ST_Slope", "HeartDisease"]

df = pd.read_csv(file_path, header=None, names=column_names)

# Chuyển cột "Name" sang kiểu chuỗi trước khi tách
df['Name'] = df['Name'].astype(str).fillna('Unknown')

# Xử lý trường hợp thiếu họ hoặc tên
df[['Firstname', 'Lastname']] = df['Name'].apply(lambda x: pd.Series(x.split(' ', 1) if ' ' in x else [x, 'Unknown']))

# Xóa cột gốc "Name"
df.drop(columns=['Name'], inplace=True)

# Hiển thị 5 dòng đầu tiên để kiểm tra
df.head()
```
